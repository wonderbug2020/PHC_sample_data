```python
#Import libraries to use
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load in the data
df = pd.read_excel('data/Preference_Policy_Homeowners.xlsx')
```

```python
#Who is the Portland Housing Center. 
    #Portland housing center provides education and financial tools to first time homebuyers. 
    #PHC targets marginalized that that have been kept from realizing the advantages of home ownership
df['Housing Client Race'].value_counts().plot(kind='pie')
```

```python
#The down payment that the national average home buyer has been putting down is 95%[1]
#Create a new column that determines how much the Mortgage to value ratio
df['CLTV'] = df['First Mortgage Amount']/df['House Purchase Price']
#Converts and cleans up this number to represent down payment as a percentage
median_down_payment = (1-round(df['CLTV'].median(),3))*100
#Prints the value. 
print(f'The median down payment is for buyers with PHC is {median_down_payment}%')
plt.bar(x=['National', 'PHC'], height=[5,25.5])
```
